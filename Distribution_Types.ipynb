```python
import pandas as pd
import numpy as np

# visualization 
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

# distributions
from scipy.stats import norm
from scipy.stats import lognorm
```

### Normal Distribution

```python
# now lets generate random variables betweem -10, 10
x = np.linspace(-10, 10, 100)

plt.figure(figsize=(8, 6))

# normal distribution takes 2 parameters: mean, std and pdf is Probability Density Function

plt.plot(norm(0, 1).pdf(x), label='Params: 0, 1')
plt.plot(norm(0, 2).pdf(x), label='Params: 0, 2')
plt.plot(norm(1, 3).pdf(x), label='Params: 1, 3')
plt.title('Standard Normal Distribution')
plt.xlabel('Values of Random variable X')
plt.ylabel('Probability')

plt.legend()
plt.show()
```

### Log-normal distribution

```python
# Varying positional arguments 
plt.figure(figsize=(8, 6))

x = np.linspace(0, 5, 100)
y1 = lognorm.pdf(x, 1, 3) 
y2 = lognorm.pdf(x, 1, 4) 
plt.plot(x, y1, "*", label='Params: 1, 3')
plt.plot(x, y2, "--", label='Params: 1, 4')
plt.title('Log-normal Distribution')
plt.xlabel('Values of Random Variable X')
plt.ylabel('Probability')
plt.legend()
plt.show()
```
